# Part 3: EDA & Visualization

```python
import seaborn as sns # For advanced visualization
import matplotlib.pyplot as plt # for plotting charts
```

```python
# 1. Line plot: Sales trend by Frequency of Purchases

df['Revenue'] = df['Total Price']  #creating new column 'Revenue' by coping 'Total Price'

# Grouping Data by 'Frequency of Purchase' and sum of revenue
freq_sales = df.groupby('Frequency of Purchases')['Revenue'].sum().sort_index()

plt.figure(figsize=(8, 5))    # for setting size of the figure for the line plot

# for creating line plot using seaborn I used sns alias
sns.lineplot(x=freq_sales.index, y=freq_sales.values, marker='o')

plt.title("Sales Trend by Frequency of Purchases")   #for writing title 

plt.ylabel("Revenue")    # for writing in y-axis or label the y-axis

plt.xlabel("Frequency of Purchases")   # label the x-axis

plt.grid(True)  # grid use for better readibility

plt.tight_layout()  # adjust layout to prevent clipping

plt.show() # for displaying the plot 
```

### Insight: 
####  1. High Revenue From Repeat Buyears means If revenue increases with higher frequency, it shows that customer that believe more in that company contribute more. This insight supports customer retention strategies. And Company can take good step for this type of customers. Customer who purchase every 3 month contribute the most revenue.

#### 2. Lower revenue: Weekly and fornightly customers generate less revenue. They can be first time buyers or not more interested to buy frequently. if they are first time buyers then company should give a good offer as possible. 

#### 3. Annually, Quarterly, and Bi-weekly buyers all show high revenue indicating strong loyalty or higher spending.

```python
# Bar chart: Top countries(Locations) by revenue
```

```python
# grouping by 'Location', sum of revenue, sort in descending order, and selecting top 10 rows
top_countries = df.groupby('Location')['Revenue'].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(10, 6))  # for setting size of the figure for the line plot

#create bar plot using seaborn
sns.barplot(data=top_countries, x='Revenue', y='Location', palette='viridis')

plt.title("Top Countries by Revenue") # setting the title of the plot

plt.xlabel("Revenue") # label the x- axis

plt.ylabel("Location") # for labeling y-axis

plt.tight_layout() # for adjust layout

plt.show() # for show the plot
```

### Insights: 
#### 1. Top revenue Generating Counrtries: In this bar char we can see that which country contribute to generate more revenue and we can also see others  

#### 2. We can more focus on that country that is generating more revenue so that company can earn more. 

### 3. If expected large economies like USA, UK, Japan etc are missing from the top 10 location it means issues in market reach higher chances. 

```python
# Pie chart: Contribution of top 5 customers
```

```python
# grouping by 'Customer ID`, sum of revenue, sort in desc and select 5 customers 
top_customers = df.groupby('Customer ID')['Revenue'].sum().sort_values(ascending=False).head(5)

explode = (0.05, 0.05, 0.05, 0.05, 0.05) #separate each slice of the pie chart for better visualization

plt.figure(figsize=(6, 6)) # for creating a figure with a specified size (6*6 inches) 

plt.pie(
    top_customers, # the revenue values of the top 5 customers
    
    labels=top_customers.index,  # for leveling (Customer IDs) for each pie slice
    
    autopct='%1.1f%%', # display percentage with 1 decimal place
    
    startangle=140, # for rotating the start of the pie chart for better layout
    
    colors=sns.color_palette('pastel'), # using pastel color palette from seaborn 
    
    explode=explode, # for slightly seperating each slice
    
    shadow = True, # adding shadow to the pie chart for a 3d effect 
    
    textprops={'fontsize': 20} ) # control font size, setting font size of labels and percentages 
    
plt.title("Top 5 Customers' Contribution to Revenue") # for title
    
plt.tight_layout() # for adjust layout
    
plt.show() # for displaying pie chart 
```

### Insights: 
#### 1. In this All 5 Customers contrubute about equally in revenue. this show revenue comes from many customers, not just a few big spenders. 

#### 2. Each Customer has only one transaction, so we can say that top sependers are not loyal. top spenders bought expensive products. 

#### 3. Company should focus on attracting many customers, not just big spenders. promote high value items with good price so that company will boost their revenue. 

```python
# Heatmap: Sales by month and country(Location)
```

```python
# grouping the data by 'Frequency of Purchase' and 'Location'
# after that sum the 'Revenue' for each combination
# after that reshape the result into a matrix using unstack, filling missing values with 0

heatmap_data = df.groupby(['Frequency of Purchases', 'Location'])['Revenue'].sum().unstack(fill_value=0)

plt.figure(figsize=(12, 6)) # for creating a figure with a specified size (12*6 inches)


sns.heatmap(   # for drawing a heatmap using the grouped data
    
    heatmap_data, # pivoting data for heatmap
    
    cmap='YlGnBu', # for coloring yellow blue and green gradient
    
    annot=False,  # by this value will not display inside the heatmap cells
    
    fmt=".0f",  # formatting round off numbers to integers if 'annot = true'
    
    linewidths=0.5) # for adding lines between cells for clarity

plt.title("Revenue Heatmap: Frequency of Purchases vs Location") #for title

plt.xlabel("Location") # for labeling the x-axis

plt.ylabel("Frequency of Purchases") # for labeling the y-axis

plt.tight_layout() # for adjusting layout to prevnt overlapping text or labels 

plt.show() # for showing heatmap
```

### Insights: 
#### 1. Revenue Distribution: The heatmap shows revenue varies by purchase frequncy and location. Some locations repeated more times means this types of location are contribution more in generation revenue.

#### 2. Locations with 'Weekly' or 'Fortnightly' showing higher revenue. that means these are active markets.

#### 3. States with 'Annually' or 'Quarterly' showing lower revenue. that means company should more focust on these locations.

```python
# Total Revenue by each Category 
```

```python
 # Grouping the data by 'Category' and calculate the total Revenue for each category

categorywise_total_revenue = df.groupby('Category')['Revenue'].sum().reset_index()
categorywise_total_revenue

# creating barplot using seaborn to show revenue by each category 
sns.barplot(data= categorywise_total_revenue, x= 'Category', y= 'Revenue', palette='viridis') #palette is collection of colors

plt.title('Total Revenue by each Category')
plt.xlabel('Category')
plt.ylabel('Revenue')

plt.tight_layout()
plt.show()
```

### Insights: 
#### 1. Clothing Category is generating highest revenue. after that accessories category is generating higher revenue. So company should more focust on this type of category. 

#### 2. footwear is generating lowest revenue. 

```python
# to see location wise revenue minimum and maximum

Location_wise_total_revenue = df.groupby('Location')['Revenue'].agg(['sum', 'max', 'min'])

# Find the location with maximum revenue
max_revenue_location = Location_wise_total_revenue['sum'].idxmax()

# highest total revenue value among all locations
max_revenue = Location_wise_total_revenue['sum'].max()

# Find the location with minimum revenue
min_revenue_location = Location_wise_total_revenue['sum'].idxmin() 

# lowest total revenue value among all locations
min_revenue = Location_wise_total_revenue['sum'].min()


# Display the results
print(f"Location with maximum revenue: {max_revenue_location} with revenue: {max_revenue}")
print(f"Location with minimum revenue: {min_revenue_location} with revenue: {min_revenue}")
```

```python
# Location wise revenue 
```

```python
# Grouping the DataFrame by 'Location' and calculate the total revenue for each location
Location_wise_total_revenue = df.groupby('Location')['Revenue'].sum().reset_index()

# Sort the locations based on total revenue in descending order (highest first)
Location_wise_total_revenue = Location_wise_total_revenue.sort_values(by='Revenue', ascending=False)

plt.figure(figsize=(15, 6))

# Creating a barplot using seaborn:
# x-axis = Location, y-axis = Revenue and using 'viridis' color palette for better visualizations
sns.barplot(data=Location_wise_total_revenue, x='Location', y='Revenue', palette='viridis')

plt.title('Total Revenue by Location', fontsize=14)

plt.xlabel('Location', fontsize=12)

plt.ylabel('Total Revenue', fontsize=12)

# Rotating x-axis labels by 45 degrees for better readability
plt.xticks(rotation=45)  

plt.tight_layout()

plt.show()
```

### Insights:  
#### 1. In this We can Which location is generating more revenue and which location is generating low revenue. 

#### 2. Montana, illinois, and california is generation more revenue. that means these are active and strong market. 


```python
# Count how many times each payment method was used and reset the index to get a DataFrame
customer_payment_by = df['Payment Method'].value_counts().reset_index()  # The new column for counts is named 'count_payment_method'

#renaming the columns to 'Payment Method' and 'Count'
customer_payment_by.columns = ['Payment Method', 'Count'] # 

# for creating barplot using seaborn 
sns.barplot(data= customer_payment_by, x= 'Payment Method', y='Count', palette='viridis') 

# for Setting the title of the plot
plt.title('Total payment count by each payment method')

plt.xticks(rotation=45)  # Rotating x-axis labels 45 degrees so they don't overlap

plt.tight_layout() # it Automatically adjust layout to fit labels and title properly

plt.show() # showing final result or bar chart 
```

### Insights: 
#### 1. Popular Payment Methods: Credi Card is most popular payment method. suggesting this is more preferred by the customers. 

#### 2. Less Common Methods: Cash and Bank transfer have lower usage, suggesting they are less preferred by the customers. 

#### 3. Company should prioritize supporting credit card so that company can earn more and generate more revenue.
